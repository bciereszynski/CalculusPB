ZADANIE 2

```python
#2.A
r,t = symbols('r t')
w=2*(x**2+y**2)*exp(-4*(x**2+y**2))
w
```

```python
x, y = symbols('x y')
plot_implicit(And(x**2 + y**2 <= 3, x**2 + y**2 >= 2), (x,-5,5), (y,-5,5), show=True, line_color='orange')
```

```python
fs=w.subs({x:r*cos(t),y:r*sin(t)})
fs
```

```python
fss=simplify(fs)
fss
```

```python
integ2 = Integral(fss*r,(t,0,2*pi),(r,sqrt(2),sqrt(3)))
integ2
```

```python
integ2.doit().evalf(10)
```

```python
#2.B
w = 1 / (x**2 + y**2 + 1)**3
w
```

```python
plot_implicit(And(x**2+y**2 <= 1, x <= 0, y >= 0), (x, -5, 5),
              (y, -5, 5), show=True, line_color='orange')
```

```python
r,t = symbols('r t')
fs=w.subs({x:r*cos(t),y:r*sin(t)})
display(fs)
fs = fs.simplify()
display(fs)
```

```python
integral = Integral(fs*r,(r,0,1),(t,pi/2,pi))
display(integral)
integral.doit().evalf()
```

```python
f1 = Eq(x, 0)
display(f1)
f2 = Eq(x, 2)
display(f2)
f3 = Eq(y + (x - 2) ** 2, 4)
display(f3)
f4 = Eq(x + y, 2)
display(f4)

p1 = plot_implicit(f1, (x, -5, 5), (y, -5, 5), show=False)
p2 = plot_implicit(f2, (x, -5, 5), (y, -5, 5), show=False)
p3 = plot_implicit(f3, (x, -5, 5), (y, -5, 5), show=False)
p4 = plot_implicit(f4, (x, -5, 5), (y, -5, 5), show=False)
p5 = plot_implicit(And(y+(x-2)**2 <= 4, x+y >= 2, x <= 2, x >= 0), (x,-5,5), (y,-5,5), show=False, line_color='orange')
p1.append(p2[0])
p1.append(p3[0])
p1.append(p4[0])
p1.append(p5[0])
p1.show()
```

```python
f = 2 - x
display(f)
g = 4 - (x - 2) ** 2
display(g)
solution = solve(Eq(f, g))[0].evalf(10)
solution
```

```python
integral = Integral(g-f, (x, solution, 2))
display(integral)
display(integral.doit())
```

```python
f = 6 - 5 * x - 3 * y
display(f)

p1 = plot_implicit(f, (x, -5, 5), (y, -5, 5), show=False)
p2 = plot_implicit(And(f >= 0, x >= 0, y>= 0), (x,-5,5), (y,-5,5), show=False, line_color='orange')

p1.append(p2[0])
p1.show()
```

```python
solution = solve(f, x)[0]
display(solution)

pole = Integral(sqrt(1 + f.diff(x)**2 + f.diff(y)**2),  (x, 0, solution), (y, 0, 2))
display(pole)
pole.doit().evalf(10)
```
