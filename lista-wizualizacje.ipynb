1. Zaproponuj 3 funkcje jednej zmiennej. Umieść je na jednym wykresie, przy czym funkcję 1
ogranicz do argumentów x[-10,-1], funkcję 2 ogranicz do argumentów x[-5,3], funkcję 3
ogranicz do argumentów x[4,10]. Każdą funkcję zaznacz innym kolorem

```python
expr1 = 3*sin(-x*2)*cos(-x/2)+exp(2/x)+10*cos(sin(1/x))
expr1
```

```python
expr2 = log(3*x+20)**2-2**(sqrt(5*x+10))
expr2
```

2. Zaproponuj funkcję „klamerkową” , np. 1+x2 dla x>5, 4 dla x=5 i -3x+4 dla x<5. Narysuj na
jednym wykresie, różnymi kolorami, tę funkcję oraz jej pochodną.

```python
expr = Piecewise((x**2+x**3-log(x**8+1), And(x >= -5,x < 5)),
(-x**3+15*exp(x), And(x <-5)),
(sqrt(x)+sin(x), And(x >= 5)))
expr_diff= diff(expr)
expr
```

```python
p2_1 = plot(expr, show=False, label='funkcja',
line_color='green')
p2_2 = plot(expr_diff, show=False, label='pochodna')
p2= plot(title="zad 2", show=False,
          xlabel="os x", ylabel="os y", legend=True)
p2.append(p2_1[0])
p2.append(p2_2[0])
p2.show()
```

przedstawia, jaki jest wzór przedstawionej powierzchni, jakie wybrano parametry.

Elipsoida to powierzchnia zamknięta, której przekrój dowolną płaszczyzną jest elipsą.
Wzór zawiera 3 parametry a, b, c. Parametr a i b odpowiadają za szerokość i długość elipsoidy. Parametr c odpowiada za jej wysokość(c do góry i -c do dołu od środka). Wybrano parametry a i b = 4 oraz parametr c =2.

```python
a,b,c = symbols("a b c")
wzor_elipsoida = x**2/a**2 + y**2/b**2+z**2/c**2 - 1
wzor_elipsoida
```

```python
from sympy.plotting import plot3d_parametric_surface

expr1 = a*cos(x)*cos(y)
expr2 = b*sin(x)*cos(y)
expr3 = c*sin(y)

p17 = plot3d_parametric_surface(expr1.subs(a,4), expr2.subs(b,4), expr3.subs(c,2),
                           (x, -pi, pi), (y, -pi/2, pi/2), 
                           xlabel='os x', ylabel='os y',zlabel='os z',
                           title='zad 17 elipsoida',Legend=True)
```
