```python
from sympy import *
init_printing(use_unicode=True)
from sympy.plotting import plot
from sympy.plotting import plot3d
x, y, z =symbols('x y z')
```

```python
#ZAD1 
#a
expr1 = x**2+y+z**3
expr1
```

```python
integ1 = Integral(expr1, (x,1,2),(y,3,4),(z,5,6) )
integ1
```

```python
integ1.doit().evalf(5)
```

```python
#zad2
f = 2 * x ** 2 + 2 * y ** 2
r, t, h = symbols('r t h')
```

```python
fs = f.subs({x: r * cos(t), y: r * sin(t), z: h})
fs
```

```python
fss = fs.simplify()
fss
```

```python
# 2.1
integral = Integral(fss * r, (r, 0, 3), (t, 0, 2 * pi), (h, 0, 4))
integral
```

```python
integral.doit()
```

```python
#3
r, fi, psi = symbols('r fi psi')
expr3 = z
expr3
```
