```python
from sympy import *
init_printing(use_unicode=True)
from sympy.plotting import plot
x, y, z, l =symbols('x y z l')
```

```python
p = plot_implicit(And(x>0, x<1-abs(y)), line_color='red',
          xlabel="os x", ylabel="os y",legend=True, show = false)
p.append(plot_implicit(Eq(x,0), show=False)[0])
p.append(plot_implicit(Eq(x,1-abs(y)), show=False)[0])
p.show()
```

```python
integ = Integral(1,(z,0,10-5*x-2*y), (y,x-1,1-x),(x,0,1))
integ
integ.doit()
```

```python
p = plot_implicit(And(x>0, y>0, 3*x+6*y<12), line_color='red',
          xlabel="os x", ylabel="os y",legend=True, show = false)
p.show()
```
