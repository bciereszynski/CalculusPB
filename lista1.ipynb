```python
x = symbols('x')
expr = x*sqrt(100-x**2)
expr
```

Dziedzina i asymptoty

```python
domain = solve(100-x**2>=0)
domain
```

```python
lim_expr = expr/x
lim_expr
```

```python
A = limit(lim_expr, x, 10)
A
```

```python
B = limit(expr-A*x, x, 10)
asym1 = A*x+B
asym1
```

```python
A = limit(lim_expr, x, -10)
A
```

```python
B= limit(expr-A*x, x, -10)
asym2 = A*x+B
asym2
```
