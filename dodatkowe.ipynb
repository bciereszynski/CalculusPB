Definiowanie zmiennych i funkcji

```python
v, w ,t = symbols('v w t')

wSceny = 0.5
vAktora = 0.8
r = 8
```

```python
Xbasic = v*t*cos(w*t)
X = Xbasic.subs(v, vAktora).subs(w, wSceny)
X 
```

```python
Ybasic = v*t*sin(w*t)
Y = Ybasic.subs(v, vAktora).subs(w, wSceny)
Y
```

ZADANIE A \
Po jakim czasie aktor opuści scenę?\
\
0.8tcos(0.5t) = x \
0.8tsin(0.5t) = y \
x^2 + y^2 = r^2 \
x = sqrt(r^2 - y^2) 

```python
time = solve(X-sqrt(r**2-Y**2))[0]
time
```

```python
time = r / vAktora
time
```

ODP: Aktor opuści scene po 10 sekundach.

Zadanie B

Ile razy obróci się scena aż aktor ją opuści?

```python
obroty = (time * wSceny)/(2*pi)
obroty.evalf(5)
```

ODP: Scena obróci się ~ 0.8 raza. Nie zrobi pełnego obrotu.

ZADANIE C \
 Jaką drogę przebędzie aktor? Oblicz długość toru ruchu aktora po 5 i 10 sekundach

```python
len = integrate(sqrt(diff(X)**2+diff(Y)**2), (t, 0, 5))
len
```

```python
len.evalf()
```

```python
len = integrate(sqrt(diff(X)**2+diff(Y)**2), (t, 0, 10))
len.evalf()
```

ODP: Aktor po 5 sek przebędzie drogę ~ 6.7 metra, a po 10 sek - ~22.25 metra

ZADANIE E \
Wykonaj odpowiedni rysunek.

```python
cX = r*cos(t)
cY = r*sin(t)
p1 = plot_parametric((X, Y), (t, 0, time), show=False)
p2 = plot_parametric((cX, cY), (t, 0, 2*pi), show=False)
p1.append(p2[0])
p1.show()
```
