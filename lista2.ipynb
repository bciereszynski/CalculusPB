```python
from sympy import *
from matplotlib import *
init_printing()
```

```python
t = symbols('t')
x = 3*(t-sin(t))
x
```

```python
y = 2*(1-cos(t))
y
```

```python
plot_parametric((x,y), (t,0,2*pi))
```

ZADANIE 1

```python
xdiff = diff(x)
xdiff
```

```python
ydiff = diff(y)
ydiff
```

```python
xydiff = ydiff/xdiff
xydiff
```

```python
xydiff.subs(t, pi/3).evalf(5)
```

```python
xydiff.subs(t, 4*pi/3).evalf(5)
```

```python
xy2diff = diff(xydiff)/xdiff
xy2diff
```

```python
xy2diff.subs(t, pi/4).evalf(3)
```

```python
xy2diff.subs(t, 5*pi/4).evalf(3)
```

```python
xy3diff = diff(xy2diff)/xdiff
xy3diff
```

```python
xy3diff.subs(t, pi/6).evalf(5)
```

```python
xy3diff.subs(t, 7*pi/6).evalf(5)
```

Zadanie 2

```python
t0 = pi / 2
x0 = x.subs(t, t0)
x0
```

```python
y0 = y.subs(t, t0)
y0
```

```python
a = (ydiff / xdiff).subs(t, t0)
a
```

```python
symbol_x = symbols('x')

styczna = a * (symbol_x - x0) + y0
styczna
```

```python
p1 = plot_parametric((x,y), (t,0,2*pi), show=False)
p2 = plot(styczna, show=False)
p1.append(p2[0])
p1.show()
```
